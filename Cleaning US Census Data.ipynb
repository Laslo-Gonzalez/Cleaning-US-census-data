# Cleaning US Census Data

## Inspect the Data!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import glob
state_files = glob.glob('states*.csv') 
state_files
```

```python
state_list = []
for filename in state_files:
    data = pd.read_csv(filename)
    state_list.append(data)

census_data  = pd.concat(state_list)
census_data
```

```python
census_data.columns
```

```python
census_data.dtypes
```

```python
census_data.Income
```

```python
#We are goint to use Regex to remove the ('$' and ',') signs
census_data.Income = census_data['Income'].replace('[\$,]', '', regex=True)
census_data.Income = census_data['Income'].replace(',', '', regex=True)
census_data.head()
```

```python
#We haver removed the signs but the Income attribute is still a object, we would like it to be a numeric
#variable, like a float
census_data.info()
```

```python
#We convert the object to a float
census_data.Income = pd.to_numeric(census_data.Income)
census_data.info()
```

```python
census_data.head()
```

```python
#Separating the GenderPop column into two different male and female columns
gender_split = census_data['GenderPop'].str.split('_')
#Creating new columns with that data
census_data['MalePop'] = gender_split.str.get(0)
census_data['FemalePop'] = gender_split.str.get(1)
```

```python
census_data.head()
```

```python
#Removing the M and F characters
census_data.MalePop = census_data['MalePop'].replace(['M'], '', regex=True)
census_data.FemalePop = census_data['FemalePop'].replace(['F'], '', regex=True)
census_data.head()
```

```python
#Converting the gender population columns into numeric variables
census_data[['MalePop', 'FemalePop']] = census_data[['MalePop', 'FemalePop']].apply(pd.to_numeric)
census_data.info()
```

```python
plt.scatter(census_data.FemalePop, census_data.Income)
```

```python
sns.scatterplot(x=census_data.FemalePop, y=census_data.Income, data=census_data)
plt.title('Scatterplot of Female Population vs Income')
plt.xlabel('Population in tens of millions')
plt.ylabel('Income in USD')
```

```python
#Checking for Missing Values
census_data.FemalePop.isna().value_counts()
```

```python
census_data.MalePop.isna().value_counts()
```

```python
#We have 3 missing values in the Female Population column and no in the male
#We can fill those missing values with
#FemalePop = TotalPop - MalePop
census_data = census_data.fillna(value={'FemalePop': census_data.TotalPop - census_data.MalePop})
```

```python
#checking to see if there are still missing values
census_data.FemalePop.isna().value_counts()
```

```python
#Checking for duplicates
census_data.duplicated(subset=['State']).value_counts()
```

```python
#Dropping Duplicates
census_data = census_data.drop_duplicates(subset=['State'])
```

```python
census_data.duplicated(subset=['State']).value_counts()
```

```python
#Plotting again without duplicates or missing values
sns.scatterplot(x=census_data.FemalePop, y=census_data.Income, data=census_data)
plt.title('Scatterplot of Female Population vs Income')
plt.xlabel('Population in tens of millions')
plt.ylabel('Income in USD')
```

## Histogram of Races

```python
census_data.columns
```

```python
census_data.isna().value_counts().reset_index()
```

```python
census_data.Pacific.isna().value_counts()
```

```python
#To plot histograms we need to change the string values to numeric, as well as fill missing values
census_data.Hispanic = census_data.Hispanic.replace('[%]', '', regex=True)
census_data.White = census_data.White.replace('[%]', '', regex=True)
census_data.Black = census_data.Black.replace('[%]', '', regex=True)
census_data.Native = census_data.Native.replace('[%]', '', regex=True)
census_data.Asian = census_data.Asian.replace('[%]', '', regex=True)
census_data.Pacific = census_data.Pacific.replace('[%]', '', regex=True)
```

```python
census_data.head()
```

```python
#COnverting to numeric
census_data[['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']] = \
census_data[['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']].apply(pd.to_numeric)
census_data.info()
```

```python
#filling the missing values
sum_some_races = (census_data.Hispanic + census_data.White + census_data.Black + census_data.Native \
+ census_data.Asian)
census_data = census_data.fillna(value = {'Pacific': 100 - sum_some_races})
census_data.Pacific.isna().value_counts()
```

```python
for race in ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']:
    plt.hist(census_data[race])
    plt.title("Histogram of the Percentage of {} People per State".format(race))
    plt.xlabel("Percentage")
    plt.ylabel("Frequency")
    plt.show()
    plt.clf()
```
